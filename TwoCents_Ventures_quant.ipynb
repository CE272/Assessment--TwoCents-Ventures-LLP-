```python
import os, json, warnings
from datetime import datetime
import numpy as np
import pandas as pd
import yfinance as yf
import optuna
import matplotlib.pyplot as plt
from dataclasses import dataclass

warnings.filterwarnings("ignore")

# ----------------------------
# CONFIG
# ----------------------------
TICKERS = ["AAPL", "MSFT", "TSLA", "BTC-USD", "ETH-USD"]
START_DATE = "2022-01-01"
INITIAL_CAPITAL = 100000.0
IS_DAYS = 365
OOS_DAYS = 90
STEP_DAYS = 90
N_TRIALS = 20
os.makedirs("data", exist_ok=True)
os.makedirs("output", exist_ok=True)


# ----------------------------
# DATA LOADER
# ----------------------------
def download_prices(tickers, start=START_DATE, end=None):
    end = end or datetime.now().strftime("%Y-%m-%d")
    data = yf.download(tickers, start=start, end=end, progress=False)["Close"]
    df = data.ffill().bfill().dropna(how="any")
    df.to_csv("data/real_assets.csv")
    print(f"Downloaded {df.shape[0]} days of data for {len(df.columns)} assets")
    return df


# ----------------------------
# BASIC CLASSES
# ----------------------------
@dataclass
class Bar:
    dt: pd.Timestamp
    symbol: str
    close: float


class DataHandler:
    def __init__(self, price_df):
        self.df = price_df
        self.dates = self.df.index
        self.symbols = self.df.columns
        self.ptr = 0

    def has_next(self):
        return self.ptr < len(self.dates)

    def next_bar(self):
        dt = self.dates[self.ptr]
        bars = [Bar(dt, s, float(self.df.loc[dt, s])) for s in self.symbols]
        self.ptr += 1
        return bars


class ExecutionHandler:
    def execute_order(self, order, price):
        return {**order, "fill_price": float(price)}


class Portfolio:
    def __init__(self, symbols, cash=INITIAL_CAPITAL):
        self.cash = cash
        self.positions = {s: 0 for s in symbols}
        self.trades = []
        self.snapshots = []

    def update_fill(self, fill):
        s = fill["symbol"]
        q = fill["quantity"] if fill["direction"] == "BUY" else -fill["quantity"]
        p = fill["fill_price"]
        self.positions[s] += q
        self.cash -= q * p
        self.trades.append(fill)

    def snapshot(self, prices, timestamp):
        value = self.cash + sum(self.positions[s] * prices[s] for s in prices)
        rec = {"timestamp": timestamp, "total_equity": value}
        rec.update({f"pos_{s}": self.positions[s] for s in self.positions})
        self.snapshots.append(rec)


# ----------------------------
# ALPHAS
# ----------------------------
class PairsAlpha:
    def __init__(self, x, y, lookback=60, z_open=2.0, z_close=0.5):
        self.x, self.y = x, y
        self.lookback, self.z_open, self.z_close = lookback, z_open, z_close
        self.hist = {x: [], y: []}
        self.pos = 0

    def on_bar(self, bars):
        prices = {b.symbol: b.close for b in bars}
        self.hist[self.x].append(prices[self.x])
        self.hist[self.y].append(prices[self.y])
        if len(self.hist[self.x]) < self.lookback:
            return []
        px = np.array(self.hist[self.x][-self.lookback:])
        py = np.array(self.hist[self.y][-self.lookback:])
        if np.any(np.isnan(px)) or np.any(np.isnan(py)):
            return []
        spread = np.log(px) - np.log(py)
        mu, sd = np.nanmean(spread), np.nanstd(spread)
        if not np.isfinite(sd) or sd < 1e-9:
            return []
        z = (spread[-1] - mu) / sd
        if not np.isfinite(z):
            return []
        orders = []
        if z > self.z_open and self.pos == 0:
            orders += [{"symbol": self.x, "quantity": 1, "direction": "SELL"},
                       {"symbol": self.y, "quantity": 1, "direction": "BUY"}]
            self.pos = -1
        elif z < -self.z_open and self.pos == 0:
            orders += [{"symbol": self.x, "quantity": 1, "direction": "BUY"},
                       {"symbol": self.y, "quantity": 1, "direction": "SELL"}]
            self.pos = 1
        elif abs(z) < self.z_close and self.pos != 0:
            if self.pos == 1:
                orders += [{"symbol": self.x, "quantity": 1, "direction": "SELL"},
                           {"symbol": self.y, "quantity": 1, "direction": "BUY"}]
            else:
                orders += [{"symbol": self.x, "quantity": 1, "direction": "BUY"},
                           {"symbol": self.y, "quantity": 1, "direction": "SELL"}]
            self.pos = 0
        return orders


class BreakoutAlpha:
    def __init__(self, symbols, lookback=20):
        self.symbols = symbols
        self.lookback = lookback
        self.hist = {s: [] for s in symbols}
        self.pos = {s: 0 for s in symbols}

    def on_bar(self, bars):
        prices = {b.symbol: b.close for b in bars}
        orders = []
        for s in self.symbols:
            self.hist[s].append(prices[s])
            if len(self.hist[s]) < self.lookback:
                continue
            window = self.hist[s][-self.lookback:]
            high, low, cur = np.max(window), np.min(window), window[-1]
            if cur > high and self.pos[s] == 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "BUY"})
                self.pos[s] = 1
            elif cur < low and self.pos[s] == 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL"})
                self.pos[s] = -1
            elif self.pos[s] != 0 and low < cur < high:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL" if self.pos[s] == 1 else "BUY"})
                self.pos[s] = 0
        return orders


class MTFMomentumAlpha:
    def __init__(self, symbols, short=7, long=30):
        self.symbols, self.short_lb, self.long_lb = symbols, short, long
        self.hist = {s: [] for s in symbols}
        self.pos = {s: 0 for s in symbols}

    def on_bar(self, bars):
        prices = {b.symbol: b.close for b in bars}
        orders = []
        for s in self.symbols:
            self.hist[s].append(prices[s])
            if len(self.hist[s]) < self.long_lb:
                continue
            short_ma = np.mean(self.hist[s][-self.short_lb:])
            long_ma = np.mean(self.hist[s][-self.long_lb:])
            if short_ma > long_ma and self.pos[s] == 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "BUY"}); self.pos[s] = 1
            elif short_ma < long_ma and self.pos[s] == 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL"}); self.pos[s] = -1
            elif abs(short_ma - long_ma) < 1e-6 and self.pos[s] != 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL" if self.pos[s] == 1 else "BUY"}); self.pos[s] = 0
        return orders


class RotationAlpha:
    def __init__(self, symbols, lookback=30, top_k=2):
        self.symbols, self.lookback, self.top_k = symbols, lookback, top_k
        self.hist = {s: [] for s in symbols}
        self.pos = {s: 0 for s in symbols}

    def on_bar(self, bars):
        prices = {b.symbol: b.close for b in bars}
        for s in self.symbols:
            self.hist[s].append(prices[s])
        if len(next(iter(self.hist.values()))) < self.lookback:
            return []
        scores = {}
        for s in self.symbols:
            arr = np.array(self.hist[s][-self.lookback:])
            rets = np.diff(arr) / arr[:-1]
            scores[s] = np.nanmean(rets) / (np.nanstd(rets) + 1e-9)
        top = {s for s, _ in sorted(scores.items(), key=lambda kv: kv[1], reverse=True)[:self.top_k]}
        orders = []
        for s in self.symbols:
            if s in top and self.pos[s] == 0:
                orders.append({"symbol": s, "quantity": 1, "direction": "BUY"}); self.pos[s] = 1
            elif s not in top and self.pos[s] == 1:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL"}); self.pos[s] = 0
        return orders


class OrderbookAlpha:
    def __init__(self, symbols, lookback=5, vol_mult=2.0):
        self.symbols, self.lookback, self.vol_mult = symbols, lookback, vol_mult
        self.hist_p = {s: [] for s in symbols}
        self.hist_v = {s: [] for s in symbols}
        self.pos = {s: 0 for s in symbols}

    def on_bar(self, bars):
        prices = {b.symbol: b.close for b in bars}
        orders = []
        for s in self.symbols:
            self.hist_p[s].append(prices[s])
            if len(self.hist_p[s]) < 2:
                continue
            ret = prices[s] / self.hist_p[s][-2] - 1
            vol_proxy = abs(ret) * 1000
            self.hist_v[s].append(vol_proxy)
            if len(self.hist_v[s]) < self.lookback:
                continue
            mean_v = np.mean(self.hist_v[s][-self.lookback:])
            if not np.isfinite(mean_v) or mean_v < 1e-9:
                continue
            cur_v = self.hist_v[s][-1]
            if cur_v > self.vol_mult * mean_v:
                sign = np.sign(ret)
                if sign > 0 and self.pos[s] == 0:
                    orders.append({"symbol": s, "quantity": 1, "direction": "BUY"}); self.pos[s] = 1
                elif sign < 0 and self.pos[s] == 0:
                    orders.append({"symbol": s, "quantity": 1, "direction": "SELL"}); self.pos[s] = -1
            elif self.pos[s] != 0 and cur_v <= self.vol_mult * mean_v:
                orders.append({"symbol": s, "quantity": 1, "direction": "SELL" if self.pos[s] == 1 else "BUY"}); self.pos[s] = 0
        return orders


# ----------------------------
# BACKTEST
# ----------------------------
def run_backtest(df, alphas):
    dh, ex, pf = DataHandler(df), ExecutionHandler(), Portfolio(df.columns)
    while dh.has_next():
        bars = dh.next_bar()
        dt = bars[0].dt
        prices = {b.symbol: b.close for b in bars}
        orders = []
        for a in alphas:
            orders.extend(a.on_bar(bars))
        for o in orders:
            o["timestamp"] = dt
            fill = ex.execute_order(o, prices[o["symbol"]])
            pf.update_fill(fill)
        pf.snapshot(prices, dt)
    eq = pd.DataFrame(pf.snapshots).set_index("timestamp")
    return pf, eq


# ----------------------------
# OPTUNA OBJECTIVES
# ----------------------------
def safe_return(eq):
    if eq.empty or np.any(np.isnan(eq["total_equity"])):
        return 999.0
    ret = eq["total_equity"].iloc[-1] / eq["total_equity"].iloc[0] - 1
    return 999.0 if not np.isfinite(ret) else -ret


def opt_pairs(trial, df):
    lb = trial.suggest_int("lookback", 20, 120)
    zo = trial.suggest_float("z_open", 1.2, 3.0)
    zc = trial.suggest_float("z_close", 0.2, 1.0)
    a = PairsAlpha("AAPL", "MSFT", lb, zo, zc)
    _, eq = run_backtest(df[["AAPL", "MSFT"]], [a])
    return safe_return(eq)


def opt_breakout(trial, df):
    lb = trial.suggest_int("lookback", 10, 60)
    a = BreakoutAlpha(["TSLA", "BTC-USD", "ETH-USD"], lb)
    _, eq = run_backtest(df[["TSLA", "BTC-USD", "ETH-USD"]], [a])
    return safe_return(eq)


def opt_mtf(trial, df):
    s = trial.suggest_int("short", 3, 15)
    l = trial.suggest_int("long", 20, 90)
    if s >= l:
        return 999.0
    a = MTFMomentumAlpha(df.columns, short=s, long=l)
    _, eq = run_backtest(df, [a])
    return safe_return(eq)


def opt_rotation(trial, df):
    lb = trial.suggest_int("lookback", 10, 60)
    tk = trial.suggest_int("top_k", 1, 3)
    a = RotationAlpha(df.columns, lb, tk)
    _, eq = run_backtest(df, [a])
    return safe_return(eq)


def opt_orderbook(trial, df):
    lb = trial.suggest_int("lookback", 3, 10)
    vm = trial.suggest_float("vol_mult", 1.5, 4.0)
    a = OrderbookAlpha(df.columns, lb, vm)
    _, eq = run_backtest(df, [a])
    return safe_return(eq)


# ----------------------------
# WFO
# ----------------------------
def walk_forward_optimization(df):
    results = []
    dates = df.index
    t = 0
    while True:
        if t + IS_DAYS + OOS_DAYS >= len(dates):
            break
        is_df = df.iloc[t:t+IS_DAYS]
        oos_df = df.iloc[t+IS_DAYS:t+IS_DAYS+OOS_DAYS]
        print(f"WFO step: IS {is_df.index[0].date()}→{is_df.index[-1].date()} | OOS {oos_df.index[0].date()}→{oos_df.index[-1].date()}")
        tuned = {}

        def safe_optimize(objective, df, defaults):
            try:
                study = optuna.create_study(direction="minimize")
                study.optimize(lambda tr: objective(tr, df), n_trials=N_TRIALS)
                if len(study.trials) == 0 or study.best_trial.state != optuna.trial.TrialState.COMPLETE:
                    return defaults
                return study.best_params
            except Exception as e:
                print("Optuna failed:", e)
                return defaults

        tuned["pairs"] = safe_optimize(opt_pairs, is_df, {"lookback": 60, "z_open": 2.0, "z_close": 0.5})
        tuned["breakout"] = safe_optimize(opt_breakout, is_df, {"lookback": 20})
        tuned["mtf"] = safe_optimize(opt_mtf, is_df, {"short": 7, "long": 30})
        tuned["rotation"] = safe_optimize(opt_rotation, is_df, {"lookback": 30, "top_k": 2})
        tuned["orderbook"] = safe_optimize(opt_orderbook, is_df, {"lookback": 5, "vol_mult": 2.0})

        alphas = [
            PairsAlpha("AAPL", "MSFT", **tuned["pairs"]),
            BreakoutAlpha(["TSLA", "BTC-USD", "ETH-USD"], **tuned["breakout"]),
            MTFMomentumAlpha(df.columns, **tuned["mtf"]),
            RotationAlpha(df.columns, **tuned["rotation"]),
            OrderbookAlpha(df.columns, **tuned["orderbook"]),
        ]

        _, eq_is = run_backtest(is_df, alphas)
        _, eq_oos = run_backtest(oos_df, alphas)

        results.append({
            "is_start": str(is_df.index[0].date()),
            "oos_start": str(oos_df.index[0].date()),
            "params": tuned,
            "is_return": None if eq_is.empty else float(eq_is["total_equity"].iloc[-1] / eq_is["total_equity"].iloc[0] - 1),
            "oos_return": None if eq_oos.empty else float(eq_oos["total_equity"].iloc[-1] / eq_oos["total_equity"].iloc[0] - 1)
        })
        t += STEP_DAYS
    return results


# ----------------------------
# MAIN
# ----------------------------
def main():
    df = download_prices(TICKERS)
    results = walk_forward_optimization(df)
    with open("wfo_results.json", "w") as f:
        json.dump(results, f, indent=2)
    print("WFO done → wfo_results.json")

    # Final run using last tuned params
    last = results[-1]["params"]
    alphas = [
        PairsAlpha("AAPL", "MSFT", **last["pairs"]),
        BreakoutAlpha(["TSLA", "BTC-USD", "ETH-USD"], **last["breakout"]),
        MTFMomentumAlpha(df.columns, **last["mtf"]),
        RotationAlpha(df.columns, **last["rotation"]),
        OrderbookAlpha(df.columns, **last["orderbook"]),
    ]
    pf, eq = run_backtest(df, alphas)
    eq.to_csv("output/final_equity.csv")
    plt.figure(figsize=(10, 4))
    plt.plot(eq["total_equity"])
    plt.title("Final Combined Equity Curve (Full Data)")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("output/final_equity.png")
    results_json = {
        "final_return_pct": float(eq["total_equity"].iloc[-1] / eq["total_equity"].iloc[0] - 1),
        "n_trades": len(pf.trades),
        "last_tuned": last
    }
    with open("results.json", "w") as f:
        json.dump(results_json, f, indent=2)
    print("Saved results.json")


if __name__ == "__main__":
    main()
```

```python
# live_connector
import time
import pandas as pd
import datetime as dt
import json
from pathlib import Path

class SandboxConnector:
    """
    Template for a live/sandbox connector.
    For now, it replays historical price data in real-time-like fashion.
    """

    def __init__(self, price_df, alphas, log_dir="logs"):
        self.price_df = price_df
        self.alphas = alphas
        self.log_dir = Path(log_dir)
        self.log_dir.mkdir(exist_ok=True)
        self.trade_log = self.log_dir / "live_trades.csv"
        self.trade_log.write_text("timestamp,symbol,alpha,signal,order_size,price\n")

    def run(self, delay=0.0):
        """Stream through data row-by-row, emulate live loop."""
        print("Starting sandbox run ...")
        equity = 100000.0
        positions = {col: 0 for col in self.price_df.columns}

        for ts, row in self.price_df.iterrows():
            timestamp = ts.strftime("%Y-%m-%d %H:%M:%S")

            for alpha in self.alphas:
                sigs = alpha.on_bar(row.to_dict())
                for sig in sigs:
                    symbol, signal = sig.get("symbol"), sig.get("signal", 0)
                    if symbol not in positions:
                        continue
                    #execution at current price
                    price = row[symbol]
                    size = signal
                    positions[symbol] += size
                    equity += -size * price

                    with self.trade_log.open("a") as f:
                        f.write(f"{timestamp},{symbol},{alpha.__class__.__name__},{signal},{size},{price}\n")

            if delay:
                time.sleep(delay)

        print(f"Sandbox run complete → {self.trade_log}")


# Usage:
if __name__ == "__main__":
    import yfinance as yf
    from alphas import PairsAlpha

    df = yf.download(["AAPL", "MSFT"], start="2024-01-01", end="2024-06-01")["Adj Close"]
    alpha = PairsAlpha("AAPL", "MSFT")
    conn = SandboxConnector(df, [alpha])
    conn.run(delay=0.01)
```

```python
# replay.py
import pandas as pd
import json

def replay_backtest(price_df, trade_log_path="logs/live_trades.csv", equity_start=100000):
    trades = pd.read_csv(trade_log_path, parse_dates=["timestamp"])
    trades = trades.sort_values("timestamp")

    equity = equity_start
    equity_curve = []
    for _, row in trades.iterrows():
        price = row["price"]
        signal = row["signal"]
        equity += -signal * price
        equity_curve.append({"timestamp": row["timestamp"], "equity": equity})

    eq_df = pd.DataFrame(equity_curve).set_index("timestamp")
    eq_df.to_csv("logs/replay_equity.csv")

    # Load sandbox equity for comparison (if logged)
    try:
        sandbox_eq = pd.read_csv("output/final_equity.csv", index_col=0, parse_dates=True)
        merged = eq_df.join(sandbox_eq, how="inner", rsuffix="_sandbox")
        diff = (merged["equity"] - merged["total_equity"]).abs().max()
        match = diff < 1e-6
    except Exception:
        diff, match = None, False

    summary = {"match": match, "max_diff": diff}
    with open("logs/replay_summary.json", "w") as f:
        json.dump(summary, f, indent=2)
    print(f"Replay complete → match={match}, max diff={diff}")
    return summary


if __name__ == "__main__":
    import yfinance as yf
    df = yf.download(["AAPL", "MSFT"], start="2024-01-01", end="2024-06-01")["Close"]
    replay_backtest(df)
```
